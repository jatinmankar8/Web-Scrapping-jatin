```python
import pandas
from datetime import datetime
from pytz import utc
import matplotlib.pyplot as plt
data = pandas.read_csv("reviews.csv", parse_dates=['Timestamp'])
```

```python
data['Day'] = data['Timestamp'].dt.date
day_avg = data.groupby(['Day']).mean()
day_avg
```

```python
plt.figure(figsize=(25,3))
plt.plot(day_avg.index, day_avg["Rating"])
```

```python
data
```

```python
data['Week'] = data['Timestamp'].dt.strftime("%Y-%U")
week_avg = data.groupby(['Week']).mean()
plt.figure(figsize=(25,3))
plt.plot(week_avg.index, week_avg['Rating'])
```

```python
data['Month'] = data['Timestamp'].dt.strftime("%Y-%m")
avg_month_allcourses = data.groupby(['Month', 'Course Name'])['Rating'].count().unstack()
avg_month_allcourses.plot(figsize=(25,10))
```

```python
data['Weekday'] = data['Timestamp'].dt.strftime('%A')
data['Dayindex'] = data['Timestamp'].dt.strftime('%w')
weekday_avg = data.groupby(['Weekday','Dayindex']).mean()
weekday_avg = weekday_avg.sort_values(['Dayindex'])
plt.figure(figsize=(25,6))
plt.plot(weekday_avg.index.get_level_values('Weekday'), weekday_avg)
```
